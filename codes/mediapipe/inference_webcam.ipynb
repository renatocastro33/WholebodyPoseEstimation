```python
import sys
import cv2

sys.path.insert(1,'../../src/')


import matplotlib.pyplot as plt
from models.mediapipe.model import mediapipe_model
from rtmlib import draw_skeleton
```

```python
model = mediapipe_model(mode_coco=True)
```

```python
# define a video capture object 
vid = cv2.VideoCapture(2) 

while(True): 
    
    ret, frame = vid.read() 
    if ret is None:
        break

    frame_rgb = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

    keypoints, scores = model.predict(frame_rgb)

    frame = draw_skeleton(frame, keypoints, scores, kpt_thr=0.5,
                                line_width=3,radius=3)
    # Display the resulting frame 
    cv2.imshow('frame', frame) 

    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

# After the loop release the cap object 
vid.release() 
# Destroy all the windows 
cv2.destroyAllWindows() 
```
