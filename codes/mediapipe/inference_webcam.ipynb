```python
import sys
import cv2
sys.path.insert(1, '../../src/')
import matplotlib.pyplot as plt
from models.mediapipe.model import mediapipe_model
```

```python
model = mediapipe_model()
```

```python
# import the opencv library 
import cv2 


# define a video capture object 
vid = cv2.VideoCapture(0) 

while(True): 
    
    ret, frame = vid.read() 
    if ret is None:
        break

    frame_rgb = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

    results = model.predict(frame_rgb)
    
    if results['pose'].pose_landmarks != None:

        print("pose len:",len(results['pose'].pose_landmarks.landmark))
    
    frame = model.draw_mediapipe(frame,results)
    # Display the resulting frame 
    cv2.imshow('frame', frame) 

    # the 'q' button is set as the 
    # quitting button you may use any 
    # desired button of your choice 
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

# After the loop release the cap object 
vid.release() 
# Destroy all the windows 
cv2.destroyAllWindows() 
```
